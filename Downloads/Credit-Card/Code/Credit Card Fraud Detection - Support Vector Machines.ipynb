# Credit Card Fraud Detection - Support Vector Machines

## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Get the Data

```python
# loading the dataset to a Pandas DataFrame
credit_card_data = pd.read_csv('creditcard.csv')
```

```python
# Drop Time column (not useful for fraud detection)
credit_card_data = credit_card_data.drop("Time", axis=1)

# Separate features and target
X = credit_card_data.drop("Class", axis=1)
y = credit_card_data["Class"]

print(f"Dataset shape: {X.shape}")
print(f"Class distribution:\n{y.value_counts()}")
print(f"Fraud ratio: {y.mean()*100:.3f}%")
```

```python
# Feature Scaling - Standardize the features
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
print("Features standardized successfully!")
```

```python
# Train Test Split — stratified to preserve fraud ratio in both sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.30, stratify=y, random_state=42
)

print(f"X_train: {X_train.shape}, X_test: {X_test.shape}")
print(f"Train fraud ratio: {y_train.mean()*100:.3f}%")
print(f"Test fraud ratio: {y_test.mean()*100:.3f}%")
```

```python
# Handle Class Imbalance with SMOTE (same approach as Decision Tree)
# sampling_strategy=0.3 avoids flooding with synthetic samples
from imblearn.over_sampling import SMOTE

smote = SMOTE(
    random_state=42,
    k_neighbors=7,
    sampling_strategy=0.3    # Oversample fraud to 30% of majority
)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

print("SMOTE Applied!")
print(f"Original training set class distribution:\n{pd.Series(y_train).value_counts()}")
print(f"\nAfter SMOTE class distribution:\n{pd.Series(y_train_smote).value_counts()}")
```

```python
# Hyperparameter Tuning with RandomizedSearchCV
from sklearn.svm import SVC
from sklearn.model_selection import RandomizedSearchCV, StratifiedKFold

param_dist = {
    'C': [0.1, 0.5, 1, 5, 10, 50],            # Regularization — lower C = more generalization
    'kernel': ['rbf', 'linear'],                 # RBF usually best for fraud detection
    'gamma': ['scale', 'auto', 0.01, 0.001],    # Controls decision boundary curvature
    'class_weight': ['balanced', None]           # Let SVM auto-adjust for imbalance
}

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

svm = SVC(probability=True, random_state=42)

random_search = RandomizedSearchCV(
    svm,
    param_dist,
    n_iter=30,
    cv=cv,
    scoring='f1',           # Optimize for fraud-class F1 directly
    n_jobs=-1,
    verbose=1,
    random_state=42
)

print("Starting RandomizedSearchCV for SVM...")
print("This may take a while due to SVM's training complexity on large datasets...")
```

```python
random_search.fit(X_train_smote, y_train_smote)

print(f"\nBest Parameters: {random_search.best_params_}")
print(f"Best CV F1 Score: {random_search.best_score_:.4f}")

svm_best = random_search.best_estimator_
```

```python
# Predictions on the FULL real-world test set (85,443 samples)
y_pred = svm_best.predict(X_test)
y_pred_proba = svm_best.predict_proba(X_test)[:, 1]
```

```python
from sklearn.metrics import (classification_report, confusion_matrix, 
                             f1_score, roc_auc_score, accuracy_score,
                             precision_recall_curve, roc_curve, auc)

print("=" * 60)
print("COMPREHENSIVE SVM MODEL EVALUATION")
print("=" * 60)
print(f"\nTest set size: {len(y_test)} samples ({(y_test==1).sum()} fraud)")
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

f1 = f1_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_pred_proba)
acc = accuracy_score(y_test, y_pred)

print(f"Key Metrics (Fraud Class):")
print(f"  F1-Score:    {f1:.4f}")
print(f"  ROC-AUC:     {roc_auc:.4f}")
print(f"  Accuracy:    {acc:.4f}")
```

```python
# Confusion Matrix
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

from sklearn.metrics import ConfusionMatrixDisplay

fig, ax = plt.subplots(figsize=(6, 5))
ConfusionMatrixDisplay.from_predictions(y_test, y_pred, display_labels=['Legit', 'Fraud'], ax=ax, cmap='Blues')
plt.title("SVM Confusion Matrix (Full Test Set)")
plt.grid(False)
plt.tight_layout()
plt.show()
```

```python
# ROC-AUC Curve and Precision-Recall Curve
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# ROC Curve
fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
roc_auc_val = auc(fpr, tpr)
axes[0].plot(fpr, tpr, color='darkorange', lw=2, label=f'SVM ROC (AUC = {roc_auc_val:.4f})')
axes[0].plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Random Classifier')
axes[0].set_xlabel('False Positive Rate')
axes[0].set_ylabel('True Positive Rate')
axes[0].set_title('ROC-AUC Curve')
axes[0].legend(loc="lower right")
axes[0].grid(alpha=0.3)

# Precision-Recall Curve
precision_vals, recall_vals, _ = precision_recall_curve(y_test, y_pred_proba)
axes[1].plot(recall_vals, precision_vals, color='blue', lw=2, label='SVM Precision-Recall')
axes[1].set_xlabel('Recall')
axes[1].set_ylabel('Precision')
axes[1].set_title('Precision-Recall Curve')
axes[1].legend(loc="best")
axes[1].grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

## Model Registry & Deployment Gate (MLOps)

```python
# ============================================================================
# PRODUCTION MODEL REGISTRY & COMPARISON LOGIC GATE
# ============================================================================
# Only deploy the new SVM model if it outperforms the production baseline

import json
import os
from datetime import datetime
from sklearn.metrics import precision_score, recall_score

BASELINE_METRICS_FILE = "svm_baseline_model_metrics.json"

# If no baseline exists, set initial baseline from first run
if not os.path.exists(BASELINE_METRICS_FILE):
    baseline_metrics = {
        'model_name': 'svm_baseline_v1',
        'timestamp': datetime.now().isoformat(),
        'f1_score': 0.75,
        'roc_auc': 0.90,
        'precision': 0.80,
        'recall': 0.70,
        'accuracy': 0.995
    }
    print("No baseline found. Initializing with default baseline metrics...")
else:
    with open(BASELINE_METRICS_FILE, 'r') as f:
        baseline_metrics = json.load(f)
    print(f"Loaded production baseline model: {baseline_metrics['model_name']}")

print("\n" + "="*70)
print("PRODUCTION BASELINE METRICS (Current Deployed Model)")
print("="*70)
for key, value in baseline_metrics.items():
    if key not in ['timestamp', 'model_name']:
        print(f"  {key.upper():20s}: {value:.4f}" if isinstance(value, float) else f"  {key.upper():20s}: {value}")

print("\nBASELINE F1-SCORE THRESHOLD:", baseline_metrics['f1_score'])
```

```python
# ============================================================================
# LOGIC GATE: Compare New SVM Model vs Production Baseline
# ============================================================================

new_model_metrics = {
    'model_name': 'svm_improved_v2',
    'timestamp': datetime.now().isoformat(),
    'f1_score': f1,
    'roc_auc': roc_auc,
    'precision': precision_score(y_test, y_pred, average='weighted'),
    'recall': recall_score(y_test, y_pred, average='weighted'),
    'accuracy': acc
}

print("\n" + "="*70)
print("NEW SVM MODEL METRICS (Candidate Model)")
print("="*70)
for key, value in new_model_metrics.items():
    if key not in ['timestamp', 'model_name']:
        print(f"  {key.upper():20s}: {value:.4f}" if isinstance(value, float) else f"  {key.upper():20s}: {value}")

# DEPLOYMENT GATE
print("\n" + "="*70)
print("DEPLOYMENT GATE: MODEL COMPARISON & DECISION")
print("="*70)

COMPARISON_METRIC = 'f1_score'
baseline_threshold = baseline_metrics[COMPARISON_METRIC]
new_model_score = new_model_metrics[COMPARISON_METRIC]
improvement = new_model_score - baseline_threshold
improvement_percent = (improvement / baseline_threshold) * 100

print(f"\nComparison Metric: {COMPARISON_METRIC.upper()}")
print(f"Baseline (Production):  {baseline_threshold:.4f}")
print(f"New Model (Candidate):  {new_model_score:.4f}")
print(f"Improvement:            {improvement:+.4f} ({improvement_percent:+.2f}%)")

print("\n" + "-"*70)
if new_model_score >= baseline_threshold:
    GATE_DECISION = "PASS - DEPLOY TO PRODUCTION"
    deployment_allowed = True
    print(f"\n  NEW MODEL PASSED: {GATE_DECISION}")
    print(f"   F1-score ({new_model_score:.4f}) >= baseline ({baseline_threshold:.4f})")
    print(f"   Deployment to production is APPROVED")
else:
    GATE_DECISION = "FAIL - REJECT & DO NOT DEPLOY"
    deployment_allowed = False
    print(f"\n  NEW MODEL FAILED: {GATE_DECISION}")
    print(f"   F1-score ({new_model_score:.4f}) < baseline ({baseline_threshold:.4f})")
    print(f"   Deployment to production is BLOCKED")
    print(f"   Required improvement: {(baseline_threshold - new_model_score):.4f} points")

print("\n" + "="*70)
print(f"FINAL DEPLOYMENT STATUS: {GATE_DECISION}")
print("="*70)
```

```python
# ============================================================================
# POST-GATE ACTIONS: Deploy or Reject
# ============================================================================

if deployment_allowed:
    print("\nDEPLOYMENT ACTIONS (Model Passed the Gate)")
    print("="*70)
    
    # Action 1: Update the Model Registry with new model metrics
    print("\n1. Updating Model Registry...")
    with open(BASELINE_METRICS_FILE, 'w') as f:
        json.dump(new_model_metrics, f, indent=2)
    print(f"   Model registry updated: {BASELINE_METRICS_FILE}")
    
    # Action 2: Save the trained SVM model
    print("\n2. Persisting model to registry...")
    import pickle
    model_path = "svm_production_model.pkl"
    with open(model_path, 'wb') as f:
        pickle.dump(svm_best, f)
    print(f"   Model saved: {model_path}")
    
    # Action 3: Save preprocessing artifacts (scaler)
    print("\n3. Saving preprocessing artifacts...")
    scaler_path = "svm_scaler_production.pkl"
    with open(scaler_path, 'wb') as f:
        pickle.dump(scaler, f)
    print(f"   Scaler saved: {scaler_path}")
    
    # Action 4: Log deployment info with full hyperparameters
    deployment_log = {
        'deployment_date': datetime.now().isoformat(),
        'previous_baseline': baseline_metrics,
        'new_model': new_model_metrics,
        'best_hyperparameters': random_search.best_params_,
        'improvement': improvement,
        'improvement_percent': improvement_percent,
        'decision': 'DEPLOYED'
    }
    with open('svm_deployment_log.json', 'w') as f:
        json.dump(deployment_log, f, indent=2, default=str)
    print(f"   Deployment log created: svm_deployment_log.json")
    
    print("\n" + "="*70)
    print("MODEL SUCCESSFULLY DEPLOYED TO PRODUCTION")
    print("="*70)
    print("\nArtifacts saved:")
    print(f"  - Model:    {model_path}")
    print(f"  - Scaler:   {scaler_path}")
    print(f"  - Metrics:  {BASELINE_METRICS_FILE}")
    print(f"  - Log:      svm_deployment_log.json")
    
else:
    print("\nREJECTION ACTIONS (Model Failed the Gate)")
    print("="*70)
    
    rejection_log = {
        'rejection_date': datetime.now().isoformat(),
        'baseline': baseline_metrics,
        'candidate_model': new_model_metrics,
        'best_hyperparameters': random_search.best_params_,
        'gap_to_threshold': (baseline_threshold - new_model_score),
        'decision': 'REJECTED',
        'reason': f'F1-Score {new_model_score:.4f} < baseline {baseline_threshold:.4f}',
        'recommendations': [
            'Try additional hyperparameter tuning (increase n_iter)',
            'Experiment with different kernels (poly, sigmoid)',
            'Adjust SMOTE sampling_strategy (try 0.2 or 0.5)',
            'Try ensemble methods (Random Forest, XGBoost)',
            'Feature engineering - create interaction features'
        ]
    }
    with open('svm_rejection_log.json', 'w') as f:
        json.dump(rejection_log, f, indent=2, default=str)
    
    print("\nMODEL REJECTED - NOT DEPLOYED")
    print(f"   Gap to meet baseline: {(baseline_threshold - new_model_score):.4f} F1 points")
    print("\nRecommendations:")
    for i, rec in enumerate(rejection_log['recommendations'], 1):
        print(f"   {i}. {rec}")
    print(f"\n   See 'svm_rejection_log.json' for details")
    print("\n" + "="*70)
```
